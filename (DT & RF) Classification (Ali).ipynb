```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv('data_cleaned.csv')
```

```python
df
```

```python
df.columns
```

```python
df.drop(columns = ['Unnamed: 0', 'team', 'name'], inplace = True)
```

```python
df
```

```python
df.drop(columns = ['position'], inplace = True)
```

```python
df
```

```python
low_threshold = df['current_value'].quantile(0.35)
mid_threshold = df['current_value'].quantile(0.75)


def categorize_price(value):
    if value < low_threshold:
        return 'Cheap'
    elif value < mid_threshold:
        return 'Good'
    else:
        return 'Expensive'

df['price_category'] = df['current_value'].apply(categorize_price)
```

```python
X = df.drop(['current_value', 'price_category'], axis=1)
y = df['price_category']

X_encoded = pd.get_dummies(X, drop_first=True)

# Splitting
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
dt_params = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

# Random Forest parameters
rf_params = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'criterion': ['gini', 'entropy']
}
```

```python
model= GridSearchCV(DecisionTreeClassifier(random_state=42), dt_params, cv=5, n_jobs=-1)
model.fit(X_train, y_train)
y_pred_dt = model.predict(X_test)

# DT Reporting
print(f'DecisionTree:\n {classification_report(y_test, y_pred_dt)}')
```

```python
RFmodel = GridSearchCV(RandomForestClassifier(random_state=42), rf_params, cv=5, n_jobs=-1)
RFmodel.fit(X_train, y_train)
y_pred_rf = RFmodel.predict(X_test)

# RF Reporting
print(f'RandomForest:\n {classification_report(y_test, y_pred_rf)}')
```
